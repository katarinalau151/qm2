```
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
% matplotlib inline
```

```
data = "TFR.csv"
dataTFR = pd.read_csv(data, index_col=0, encoding='latin1')
dataTFR.head()
```

```
sns.set(font_scale=1.4)
dataTFR['2018'].plot(kind='hist', figsize=(10, 10), bins=25);
plt.xlabel("Total Fertility Rate", labelpad=14)
plt.ylabel("Frequency (No. of Countries)", labelpad=14)
plt.title("Distribution of World TFR 2018", y=1.015, fontsize=22);
```

```
#Can set column to specified year to get create visualisation, e.g. 2017 
sns.set(font_scale=1.4)
dataTFR['2017'].plot(kind='hist', figsize=(10, 10), bins=25);
plt.xlabel("Total Fertility Rate", labelpad=14)
plt.ylabel("Frequency (No. of Countries)", labelpad=14)
plt.title("Distribution of World TFR 2017", y=1.015, fontsize=22);
```

```
#Upload Africa TFR Data 
from google.colab import files
uploaded = files.upload()
```

```
data2 = "TFR Africa.csv"
dataTFRAfrica = pd.read_csv(data2, index_col=0, encoding='latin1')
dataTFRAfrica.head()
```

```
dataTFRAfrica['2018'].plot(kind='hist', figsize=(10, 10), bins=25);
plt.xlabel("Distribution of TFR", labelpad=14)
plt.ylabel("Frequency (No. of Countries)", labelpad=14)
plt.title("Distribution of TFR (Africa) 2018", y=1.015, fontsize=22);
```

```
plt.figure(figsize=(10,10))
plt.hist(dataTFR['2018'], bins=25, alpha=0.8, label="World TFR")
plt.hist(dataTFRAfrica['2018'], bins=25, alpha=0.7, label="Africa TFR")
plt.xlabel("Total Fertility Rate", labelpad=14)
plt.ylabel("Frequency (No. of Countries)", labelpad=14)
plt.title("Distribution of TFR in World & Africa")
plt.legend(loc='upper right')
plt.savefig("Overlapped TFR Histogram.png")
```
